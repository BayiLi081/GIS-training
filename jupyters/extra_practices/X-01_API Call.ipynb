## <a id='toc1_2_'></a>[Nested data structure](#toc0_)

Dictionary (response_json) --> Dictionary (target_results; key: result) --> Data, Format: List (target_results; key: records)

```python
response_json['result']['records']
```

````python
import pandas as pd

target_df = pd.DataFrame(response_json['result']['records'])

target_df.head(5)
````

````python
len(target_df)
````

We can see that the API has a limitation that each call will only return 100 records.

There are two options:

1. use `&offset` + loopover the API Query

2. Just use the read-to-download CSV downloading option

````python
dataset_id = "d_3c55210de27fcccda2ed0c63fdd2b352"

url = "https://data.gov.sg/api/action/datastore_search?resource_id="  + dataset_id + "&offset=1362"
        
response = requests.get(url)
pprint(response.json())
````

````python
collection_id = 2          
url = "https://api-production.data.gov.sg/v2/public/api/collections/{}/metadata".format(collection_id)
        
response = requests.get(url)
pprint(response.json())
````
