# Transform Shapefile into KML and imported into Google Map

```python
import geopandas as gpd
```

```python
# Read a shapefile
shapefile_path = "data/SG_MRT/MRT_LRT_Stations.shp"
shapefile_data = gpd.read_file(shapefile_path)

print(shapefile_data.head())
```

```python
# divided PT_CODE into a list by the delimiter "/"
shapefile_data['PT_CODE_List'] = shapefile_data['PT_CODE'].str.split('/')
```

```python
# get all the pt code in the column PT_CODE
all_pt_code = []
for pt_code in shapefile_data['PT_CODE_List']:
    all_pt_code.extend(pt_code)
```

```python
import re

# remove numbers in each code in the list
mrt_codes = []
for code in all_pt_code:
    mrt_codes.append(re.sub(r'[0-9]', '', code))

# remove duplicates
mrt_codes = list(set(mrt_codes))
```

```python
list_mrt_gpd = dict()

for mrt_code in mrt_codes:
    mrt_gpd = shapefile_data[shapefile_data['PT_CODE'].str.contains(mrt_code)]
    list_mrt_gpd[mrt_code] = mrt_gpd
    # save the data as kml with the name of the mrt_code
    mrt_gpd['index'] = mrt_gpd.index
    # save as csv
    mrt_gpd.to_csv(f"data/SG_MRT/{mrt_code}.csv", index=False)
    mrt_gpd.to_file(f"data/SG_MRT/{mrt_code}.kml", driver='KML')
```
