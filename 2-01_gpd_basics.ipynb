### GeoDataFrame

> A GeoDataFrame object is a pandas.DataFrame that has one or more columns containing geometry.

References:

1. https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.html#geopandas-geodataframe

Step 0: Import the libraries

```python
import geopandas as gpd
import pandas as pd
```

#### Creating a point collection

```python
from shapely.geometry import Point
```

```python
pt = {'col1': ['name1'], 'geometry': [Point(1, 2)]}
pt_gdf = gpd.GeoDataFrame(pt, crs="EPSG:4326")
pt_gdf
```

```python
pt_gdf.plot()
```

Task: Add another point (Point(2, 1)) to the point collection

*Because GeoDataFrame is an extended format of DataFrame, we can use function within pandas to append the GeoDataFrame*

```python
# Add another point (Point(2, 1)) to the point collection pt_gdf
pt_gdf = pd.concat([pt_gdf, gpd.GeoDataFrame({'col1': ['name2'], 'geometry': [Point(2, 1)]}, crs="EPSG:4326")])
pt_gdf
```

Simply plot the point use the plot function

```python
pt_gdf.plot()
```

#### Create a line collection

```python
from shapely.geometry import LineString
```

Create a line from coordinates of two points

```python
line = {'col1': ['name1'], 'geometry': [LineString([(1, 2), (2, 1)])]}

line_gdf = gpd.GeoDataFrame(line, crs="EPSG:4326")

line_gdf
```

```python
line_gdf.plot()
```

#### Create a polygon collection

```python
from shapely.geometry import Polygon
```

```python
d = {'col1': ['name1'], 'geometry': [Polygon([(1, 2), (2, 1), (2, 2)])]}

gdf = gpd.GeoDataFrame(d, crs="EPSG:4326")

gdf
```

```python
gdf.plot()
```
